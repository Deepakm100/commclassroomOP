### Here we gonna deal with the text data which we need to convert it in numerical data , also for ml model we choose Logistic Regression since the problem type is binary classiification.

```python
# Importing dependencies
import numpy as np
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Downloading stopwords
import nltk
nltk.download('stopwords')
```

```python
print(stopwords.words('english'))
```

```python
# Data collection
news_dataset = pd.read_csv('train.csv')
```

```python
news_dataset.head()
```

```python
# Data preprocessing
news_dataset.isnull().sum()
```

```python
news_dataset = news_dataset.fillna('')
```

```python
# Now for our analysis prupose we take author and title of the news since taking text take lot of a time
news_dataset['content'] = news_dataset['author'] + " " + news_dataset['title']
news_dataset.head()
```

Stemming -> Reducing the word to its root form
e.g actor, acting, actress --> act

```python
port_stem = PorterStemmer()
```

```python
# defining a function
def stemming(content):
  stemmend_content = re.sub('[^a-zA-z]',' ',content)
  # --> here we replace all the things in content which are not alphabets to a empty string
  stemmend_content = stemmend_content.lower()
  stemmend_content = stemmend_content.split()
  # by using split we are making an list
  # stemming
  stemmend_content = [port_stem.stem(word) for word in stemmend_content if not word in stopwords.words('english')]
  stemmend_content = ' '.join(stemmend_content)
  # .join uses after split so that it make sure
  return stemmend_content
```

```python
news_dataset['content'] = news_dataset['content'].apply(stemming)
```

```python
# seprating the data
X = news_dataset['content'].values
Y = news_dataset['label'].values
X.shape
# Note --> by using .values me make a list of our data
# ------------------------------------------- Important tips -----------------------------------------
# Snippet 1: X and Y are pandas.Series objects. When you extract a column from a DataFrame without using .values, it remains a Series.
# Snippet 2: X and Y are NumPy arrays. Using .values converts the Series into NumPy arrays.

# Index Preservation:

# Series: When using Series, the index is preserved, which can be useful for aligning data, joining tables, etc.
# NumPy Arrays: The index information is lost when converting to NumPy arrays. They are essentially just arrays of values without the associated index.
# When to Use Which:
# Use Series (Snippet 1) when you need to perform data manipulation, analysis, and operations that leverage Pandas' powerful features and when you need to maintain the index of the data.
# Use NumPy Arrays (Snippet 2) when you are focused on numerical computations, need higher performance for mathematical operations, and do not need the additional functionalities provided by Pandas Series
```

```python
# # seprating the data
# X = news_dataset['content']
# Y = news_dataset['label']
#  here there is just simple extraction of column
```

```python
# Vectorizer
vectorizer = TfidfVectorizer()
vectorizer.fit(X)
X = vectorizer.transform(X)
X.shape
```

```python
# splitting the data into training and test data
X_train, X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

Model --> Logistic regression

```python
model = LogisticRegression()
model.fit(X_train,Y_train)
```

Evaluation --> accuracy score

```python
Y_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train_prediction,Y_train)
print(training_data_accuracy)
```

```python
# Test data evaluation
Y_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test_prediction,Y_test)
print(test_data_accuracy)
```

Making an predictive system

```python
X_new = X_test[1]
prediction = model.predict(X_new)
if prediction[0] == 0:
  print("The news is real")
else:
  print("The news is fake")
```
